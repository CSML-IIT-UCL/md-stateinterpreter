# Load data

The file COLVAR contains absolute positions of the heavy atoms (excluding hydrogens) of alanine dipeptide. They are 10 atoms, and the related columns are named: p_ID.x/y/z where ID is the particle number.

### Example of dataset import

```python
import pandas as pd
import numpy as np

atom_ids=[2, 5, 6, 7, 9, 11, 15, 16, 17, 19] #if necessary

filename='COLVAR'
headers = pd.read_csv(filename,sep=' ',skipinitialspace=True, nrows=0).columns[2:]  
df = pd.read_csv(filename,sep=' ',skipinitialspace=True, header=None,skiprows=1,names=headers,comment='#') 

#select subset of columns based on names
pos = df.filter(regex='p_')

#save features names and np array
names = pos.columns
array = pos.values

print('names:',names.shape)
print('array:',array.shape)
```

#### Labelling

```python
#label states based on whether phi < 0 or phi >0 
phi = df['phi']
state = np.zeros_like(phi)
state[ (phi > 0 ) & (phi <2.5)] = 1 
df['state'] = state

#save position for state A (0) and B (1)
posA = df.loc[df['state']==0].filter(regex='p_')
posB = df.loc[df['state']==1].filter(regex='p_')

print('A:',posA.shape)
print('B:',posB.shape)
```
